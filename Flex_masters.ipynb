```python
import pandas as pd
```

```python
df = pd.read_json('voice.json')
```

```python
df.head()
```

```python
df['createdAt'] = pd.to_datetime(df['createdAt'], unit = 'ms')
```

```python
df['Dates'] = pd.to_datetime(df['createdAt']).dt.date
df['Time'] = pd.to_datetime(df['createdAt']).dt.time
```

```python
df.drop('createdAt', axis = 1)
```

```python
column_names = ['Dates','Time','location', 'networkInfo', 'keyPerformanceIndicators', 'device', 'manual' ]
df = df.reindex(columns = column_names)
```

```python
df = df.sort_values('Dates')
```

```python
df = df.reset_index(drop = True)
```

```python
count_table = df.groupby('Dates')['Dates'].count()
```

```python
count_table = count_table.to_frame(name = 'Count')
```
